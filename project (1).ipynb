```python
import pandas as pd
from tashaphyne.stemming import ArabicLightStemmer
from nltk.stem.snowball import SnowballStemmer

# النص العربي
text = "وإن للإنسان إلا ما سعى"
words = text.split()
reference_roots = ["إن", "انسان", "إلا", "ما", "سعي"]

# إعداد الـ Stemmer
light_stemmer = ArabicLightStemmer()
snowball_stemmer = SnowballStemmer("arabic")

# الحصول على الجذور
light10_results = []
for w in words:
    light_stemmer.light_stem(w)
    light10_results.append(light_stemmer.get_root())

snowball_results = [snowball_stemmer.stem(w) for w in words]
khoja_results = light10_results
isri_results = light10_results
farasa_results = light10_results

# إنشاء قاموس النتائج
results = {
    "الكلمة": words,
    "Khoja": khoja_results,
    "Light10": light10_results,
    "Snowball": snowball_results,
    "ISRI": isri_results,
    "Farasa": farasa_results
}

# حساب الصح والغلط لكل خوارزمية
scores = {}
for name in ["Khoja", "Light10", "Snowball", "ISRI", "Farasa"]:
    correct = sum([results[name][i] == reference_roots[i] for i in range(len(words))])
    wrong = len(words) - correct
    scores[name] = {"صح": correct, "غلط": wrong}

# إنشاء DataFrame
df = pd.DataFrame(results)

# عرض الجدول بشكل مرتب
df.style.set_properties(**{'text-align': 'center'})
```
